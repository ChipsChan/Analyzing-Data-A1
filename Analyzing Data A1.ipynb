```python
# use a dictionary to store the word frequency of each text
word_frequencies_per_text = {}
total_word_frequencies = Counter()
```

```python
# iterate through each file and calculate word frequencies
for file_name in file_names:
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read()
        sentences = sent_tokenize(text)
        words = []
        for sentence in sentences:
                tokenized_words = word_tokenize(sentence)
                filtered_words = [word.lower() for word in tokenized_words if word.isalpha()]
                words.extend(filtered_words)
        word_freq = Counter(words)
        word_frequencies_per_text[file_name] = word_freq
        total_word_frequencies.update(word_freq)
```

## **Part 1.2**

```python
# initialize stemmer
porter = PorterStemmer()
lancaster = LancasterStemmer()
```

```python
# initialize dictionaries to save word frequencies
porter_word_frequencies_per_text = {}
lancaster_word_frequencies_per_text = {}
```

```python
# tokenize sentences and words
for file_name in file_names:
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read().strip()
    sentences = sent_tokenize(text)
    all_words = []
    for sentence in sentences:
        tokenized_words = word_tokenize(sentence)
        filtered_words = [word.lower() for word in tokenized_words if word.isalpha()]
        all_words.extend(filtered_words)

    # stem words
    porter_stemmed_words = [porter.stem(word) for word in all_words]
    lancaster_stemmed_words = [lancaster.stem(word) for word in all_words]
    
    porter_word_frequencies_per_text[file_name] = Counter(porter_stemmed_words)
    lancaster_word_frequencies_per_text[file_name] = Counter(lancaster_stemmed_words)
```

```python
for file_name in file_names:
    # plot histograms for Porter Stemmer
    top_words_porter = porter_word_frequencies_per_text[file_name].most_common(25)
    words_porter, counts_porter = zip(*top_words_porter)

    print(f"Porter Stemmer - Top 25 words in {file_name}:")
    for word, count in top_words_porter:
        print(f"{word}: {count}")

    plt.figure(figsize=(10,5))
    plt.bar(words_porter, counts_porter)
    plt.xticks(rotation=45)
    plt.title(f"Porter Stemmer - Top 25 words in {file_name}")
    plt.show()

    # plot histograms for Lancaster Stemmer
    top_words_lancaster = lancaster_word_frequencies_per_text[file_name].most_common(25)
    words_lancaster, counts_lancaster = zip(*top_words_lancaster)

    print(f"Lancaster Stemmer - Top 25 words in {file_name}")
    for word, count in top_words_lancaster:
        print(f"{word}: {count}")

    plt.figure(figsize=(10,5))
    plt.bar(words_lancaster, counts_lancaster)
    plt.xticks(rotation=45)
    plt.title(f"Lancaster Stemmer - Top 25 words in {file_name}")
    plt.show()
```

```python
# compare unstemmed and stemmed word frequencies
for file_name in file_names:
    print(f"\n Comparing word frequency changes in {file_name}:\n")

    # get unstemmed words frequencies
    top_unstemmed = word_frequencies_per_text[file_name].most_common(25)
    words_unstemmed, counts_unstemmed = zip(*top_unstemmed)

    # get stemmed words frequencies after Porter stemming
    top_porter = porter_word_frequencies_per_text[file_name].most_common(25)
    words_porter, counts_porter = zip(*top_porter)

    # get stemmed words frequencies after Lancaster stemming
    top_lancaster = lancaster_word_frequencies_per_text[file_name].most_common(25)
    words_lancaster, counts_lancaster = zip(*top_lancaster)

    # pring the answer of comparison
    print(" Before Stemming (Unstemmed):")
    print(", ".join(words_unstemmed))

    print("\n After Porter Stemming:")
    print(", ".join(words_porter))

    print("\n After Lancaster Stemming:")
    print(", ".join(words_lancaster))

    print("\n" + "="*50)
```

```python
# print vocabulary size in each text file
for file_name in file_names:
    original_vocab_size = len(word_frequencies_per_text[file_name])
    porter_vocab_size = len(porter_word_frequencies_per_text[file_name])
    lancaster_vocab_size = len(lancaster_word_frequencies_per_text[file_name])

    print(f"\n Vocabulary size in {file_name}:")
    print(f" Before Stemming: {original_vocab_size} unique words")
    print(f" After Porter Stemming: {porter_vocab_size} unique words")
    print(f" After Lancaster Stemming: {lancaster_vocab_size} unique words")
    print("\n" + "="*50)
```

# **Comparison of Unstemmed and Stemmed Results**


## **1. Vocabulary Size Reduction**
Observing the results above, we can see that stemming significantly reduces the number of unique words in the text.  
This occurs because different word forms are converted into their base form.  
As a result, stemming helps normalize vocabulary and makes text processing more efficient.

## **2. Changes in Top Words**
After stemming, nearly half of the top words change.  
Since different word variations are reduced to their root form, some words become more dominant in the frequency ranking.  
This transformation can be useful in search engines and NLP applications, as it groups similar words together.

## **3. Over-Stemming and Under-Stemming**

- **Over-Stemming**: Occurs when a word is reduced too aggressively, losing its original meaning.
  - Example: "was"  "wa"
  
- **Under-Stemming**: Occurs when stemming fails to reduce words to a common root, keeping unnecessary details.  
  - Example: "didn't"  "didn"



    




# **Comparison of Porter and Lancaster Stemming Results**


## **1. Top Words Comparison**
When comparing the top words before and after stemming, we observe that the results of the Porter Stemmer and Lancaster Stemmer are more similar to each other than to the original (unstemmed) word list.  
This is because both stemmers aim to reduce words to their root forms, causing common variations to merge.

## **2. Vocabulary Size Differences**
The Lancaster Stemmer produces a smaller vocabulary size than the Porter Stemmer.  
Additionally, as the length of the text increases, the difference in vocabulary size between the two stemmers becomes more significant.  
This suggests that Lancaster is more aggressive in reducing unique words, leading to greater compression.

## **3. Lancaster Stemming is More Aggressive**
We can see Lancasters aggressiveness reflected in both vocabulary size reduction and top word transformations:
  - **Vocabulary Size**: Lancaster consistently results in **fewer unique words** compared to Porter.
  - **Word Transformations**: Lancaster over-stemming more frequently than Porter.  
     Example: "doctor"  "doct" (Lancaster) vs. "doctor"  "doctor" (Porter).
        This means Lancaster Stemmer sacrifices more detail, making it less suitable for tasks requiring precise word meanings.

```python
# load the spaCy language model
nlp_en = spacy.load("en_core_web_sm")  # for English
nlp_de = spacy.load("de_core_news_sm")  # for German
nlp_nl = spacy.load("nl_core_news_sm")  # for Dutch

# read the file
file_paths = {
    "English": "pg74.txt",
    "Dutch": "pg18381.txt",
    "German": "pg30165.txt"
}

# process text
def clean_text(text):
    # clean the text, remove the copyright information and directory, and extract the main body part
    text = re.sub(r"\*\*\* START OF THE PROJECT GUTENBERG EBOOK.*?\*\*\*", "", text, flags=re.DOTALL)
    text = re.sub(r"CONTENTS.*?CHAPTER\s+I", "CHAPTER I", text, flags=re.DOTALL | re.IGNORECASE)
    match = re.search(r"(CHAPTER I\n|HOOFDSTUK I|Erstes Kapitel)", text, re.IGNORECASE)
    if match:
        text = text[match.start():] 
    end_match = re.search(r"(END OF THE PROJECT GUTENBERG EBOOK.*)", text, re.IGNORECASE)
    if end_match:
        text = text[:end_match.start()]
    text = text.strip()
    return text 

cleaned_texts = {}
for language, file_path in file_paths.items():
    with open(file_path, 'r', encoding='utf-8') as file:
        raw_text = file.read()
        cleaned_text = clean_text(raw_text)
        cleaned_texts[language] = cleaned_text

# test the result of clean
for language, text in cleaned_texts.items():
    print(f"\n ~ {language} Text Preview:\n{text[:200]}...\n")
    print(f"\n ~ {language} Text Last 200 Characters:\n{text[-200:]}\n")
```

```python
import pandas as pd
```

```python
# set a language model dictionary
nlp_models = {
    "English": nlp_en,
    "Dutch": nlp_nl,
    "German": nlp_de
}

# process text and perform POS tagging
def process_text_limited(text, nlp, limit=5000):
    doc = nlp(text[:limit]) 
    pos_counts = Counter(token.pos_ for token in doc)
    return pos_counts

# calculate POS-tags statistics
pos_counts_dict = {}
for lang, path in file_paths.items():
    with open(path, "r", encoding="utf-8") as f:
        raw_text = f.read()
        cleaned_text = clean_text(raw_text)
        pos_counts_dict[lang] = process_text_limited(cleaned_text, nlp_models[lang])  

# plot a POS distribution histogram
def plot_pos_distribution(pos_counts, language):
    labels, values = zip(*pos_counts.items())
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values)
    plt.xlabel("POS Tag")
    plt.ylabel("Frequency")
    plt.title(f"POS Distribution in {language}")
    plt.xticks(rotation=45)
    plt.show()

for lang, pos_counts in pos_counts_dict.items():
    plot_pos_distribution(pos_counts, lang)

# generate POS statistics table
df_pos_counts = pd.DataFrame.from_dict(pos_counts_dict, orient="index").fillna(0).astype(int)
print(df_pos_counts)
```
